```python
from pyspark import SparkContext
```

```python
# StreamingContext is the main entry point for Spark Streaming functionality
# provides methods used to create DStreams from input sources.
from pyspark.streaming import StreamingContext
```

```python
# We are using the Spark Streaming library
# Create a local StreamingContext with two working threads 
# and batch interval of 1 second
# The second parameter "LightOnCount" is the name of the application
sc = SparkContext("local[2]", "LightOnPredict")
ssc = StreamingContext(sc, 1)
```

```python
# Create a DStream that will connect to hostname:port
# this DStream is a receiver for our host
# it receives the input
# the input is broadcasted by the Netcat command:
# nc -lk 9999 < iot_telemetry_data.csv
# this command transmits the content of iot_telemetry_data.csv ONE LINE AT A TIME 
# over port 9999
lines = ssc.socketTextStream("localhost", 9999)
```

```python
type(lines)
#pyspark.streaming.dstream.DStream
```

```python
#splt the lines which is a srting into words
#words = lines.flatMap(lambda line:line.split(','))
words = lines.flatMap(lambda line:line)
```

```python
# create the tupples
pairs = words.map(lambda word:(word,1))
```

```python
# reduce
word_counts = pairs.reduceByKey(lambda num1, num2:num1+num2)
#pyspark.streaming.dstream.TransformedDStream
```

```python
# print as the words count themselves
words.pprint()
word_counts.pprint()
```

```python
# right after running this we should run the Netcat command in a different Terminal:
# nc -lk 9999 < iot_telemetry_data.csv
# start
ssc.start()             # Start the computation

##result
# -------------------------------------------
# Time: 2021-02-24 00:56:55
# -------------------------------------------
# ('aaa', 3)
# ('', 2)
# ('oo', 1)
# ('ddd', 5)
# ('cccc', 2)
# ('aaass', 1)
# ('uuu', 1)
```
