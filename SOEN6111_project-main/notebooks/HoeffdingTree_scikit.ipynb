```python
from sklearn.model_selection import train_test_split
```

```python
from skmultiflow.trees import HoeffdingTree, HoeffdingTreeClassifier
```

```python
# prepare the data set
import pandas as pd
df=pd.read_csv("iot_telemetry_data.csv")
```

```python
#explore the data frame
df.info()
```

```python
# drop the duplicates and keep the first occurence only, modify the dataframe
df.drop_duplicates(inplace=True, keep="first")
```

```python
#convert the True/False values to 1/0
df["light"]=df["light"].astype(int)
df["motion"]=df["motion"].astype(int)
```

```python
#select the features and the target variable
#we exclude the timestamp
#?Question? Do we need to add the device?
feature_cols = ['co', 'humidity', 'lpg', 'motion','smoke','temp']
X = df[feature_cols] # Features
y = df['light'] # Target label
```

```python
#split dataset in features and target label
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
X_train
```

```python
'''
    def partial_fit(self, X, y, classes=None, sample_weight=None):
        """ Incrementally trains the model. Train samples (instances) are
        composed of X attributes and their corresponding targets y.
        Parameters
        ----------
        X: numpy.ndarray of shape (n_samples, n_features)
            Instance attributes.
        y: array_like
            Classes (targets) for all samples in X.
        classes: numpy.array
            Contains the class values in the stream. If defined, will be used
            to define the length of the arrays returned by `predict_proba`
        sample_weight: float or array-like
            Samples weight. If not provided, uniform weights are assumed.'''
```

```python
# train the Hoeffding Tree on the entire data set (not incrementally)
ht_full = HoeffdingTreeClassifier()
```

```python
#convert dataframe to numpy array
print(df.to_numpy())
```

```python
# train the Hoeffding Tree by using the entire training set
ht_full.fit(X_train.to_numpy(), y_train.to_numpy())
```

```python
#Predict the class label for test dataset
y_pred = ht_full.predict(X_test.to_numpy())
```

```python
print(y_pred)
```

```python
# What is the metrics for this classifier?
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score
```

```python
# Accuracy by comparing actual testset values and predicted values
type(y_test)
#pandas.core.series.Series
type(y_pred)
#numpy.ndarray
print(y_test.to_numpy())

print("Accuracy (test set):", accuracy_score(y_test.to_numpy(), y_pred))
# Accuracy (test set):  0.986931572777195
```

```python
'''
 score(self, X, y, sample_weight=None)[source]

    Returns the mean accuracy on the given test data and labels.

    In multi-label classification, this is the subset accuracy which is a harsh metric since you require for each sample that each label set be correctly predicted.

    Parameters

        Xarray-like, shape = (n_samples, n_features)

            Test samples.
        yarray-like, shape = (n_samples) or (n_samples, n_outputs)

            True labels for X.
        sample_weightarray-like, shape = [n_samples], optional

            Sample weights.

'''
```

```python
ht_full_mean_accuracy = ht_full.score(X_test.to_numpy(), y_test.to_numpy())
```

```python
#Note same as accuracy above
print(ht_full_mean_accuracy)
```

```python
# Accuracy on the training set
y_pred_train = ht_full.predict(X_train.to_numpy())
print("Accuracy (training set):", accuracy_score(y_train.to_numpy(),y_pred_train))
#Accuracy (training set): 0.9862927906804552
```

```python
print("F1 score (test set):", f1_score(y_test.to_numpy(), y_pred))
# F1 score  (test set):  0.9760304203164256
```

```python
print("Recall score (test set):", recall_score(y_test.to_numpy(), y_pred))
# Recall score (test set):  0.9594606621573514
```
